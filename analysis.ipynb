```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
df.head(10)
```

```python
print(df['Category'].unique())
print(df['Category'].nunique())
```

```python
print(df['Item Purchased'].unique())
```

```python
# Fill missing values in 'Review Rating' with the median rating for each 'Item Purchased'

df['Review Rating'] = df.groupby('Item Purchased')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns
```

```python
#edit column names to remove spaces and make them lowercase

df.columns = df.columns.str.replace(' ', '_').str.lower()
df.columns
```

```python
#edit column name
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
#create a new column for age group from 'age' column using qcut

labels = ['Young Adult', 'Adult', 'Middle Aged', 'Elderly']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)


```

```python
#show just age and age_group colummn
df[['age', 'age_group']].head(10)
```

```python
#Min and max values from age
print(df['age'].min())
print(df['age'].max())
```

```python
df['frequency_of_purchases'].unique()
```

```python
#new column  for frequency of purchases per day in integer values according to the frequency of purchases column    

frequency_mapping = {
    'Fortnightly'   : 14, 
    'Weekly'        : 7,   
    'Annually'      : 365, 
    'Quarterly'     : 90,
    'Bi-Weekly'     : 14,
    'Monthly'       : 30,
    'Every 3 Months': 90
}

df['frequency_days_per_perchase'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_of_purchases', 'frequency_days_per_perchase']].head(10)
```
